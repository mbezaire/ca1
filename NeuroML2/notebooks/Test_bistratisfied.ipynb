```python
from helper import input_currents
cell_file = "../cells/bistratified.cell.nml"

input_currents(cell_file,
               custom_amps_nA=[-0.3, -0.25, -0.2 -0.15, -0.1, -0.05, 0, 0.45],
               analysis_duration=1000,
               pre_zero_pulse=500,
               post_zero_pulse=500)
```
